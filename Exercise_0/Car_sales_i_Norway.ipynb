```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("../data/norway_new_car_sales_by_month.csv")

print("Columns:", df.columns.tolist())
print("Shape:", df.shape)
print("\nData types and non-null counts:")
print(df.info())
print("\nDescriptive statistics (numeric columns):")
print(df.describe())
```

```python
sum_by_year = df.groupby('Year')['Quantity'].sum().reset_index()
print(sum_by_year)
```

```python
year_plot = sum_by_year.plot(x = "Year",
             y = "Quantity",
             title = "Total Quantity by yaer",
             kind = "bar"); 

year_plot.spines["top"].set_visible(False)
year_plot.spines["right"].set_visible(False)
year_plot.legend().set_visible(False)
```

```python
sum_Co2_year = df.groupby('Year')['Avg_CO2'].sum()

Co2_year_plot = sum_Co2_year.plot(x = "Year",
             y = "Avg_CO2",
             title = "Total CO2 per year",
             kind = "bar"); #use semicolon to supress text info, which is written out by dunder __repr__ method

year_plot.spines["top"].set_visible(False)
year_plot.spines["right"].set_visible(False)
year_plot.legend().set_visible(False)
```
