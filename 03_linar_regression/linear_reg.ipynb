```python
##2.Feature scaling

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
type(scaler)
```

```python
scaler
```

```python
scaler.fit(X_train)
scaled_X_train = scaler.transform(X_train)
scaled_X_test = scaler.transform(X_test)

print(f"{scaled_X_train.min() = }")
print(f"{scaled_X_train.max() = }")
print(f"{scaled_X_test.min() = }")
print(f"{scaled_X_test.max() = }")
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
   X, y, test_size=0.33, random_state=42
   )

print(f"{X_train.shape=}")
print(f"{y_train.shape=}")
print(f"{X_test.shape=}")
print(f"{y_test.shape=}")

```

```python
scaled_X_train[:5]
```

```python
type(scaled_X_train)
```

```python
##LINEAR REGRESSION

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model 
```

```python
model.fit(scaled_X_train, y_train)
print(f"Parameters or weights: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

```python
##4.Prediction

sample_features = scaled_X_test[0].reshape(1, -1)
sample_features
```

```python
model.predict(sample_features)
```

```python
y_test.iloc[0]
```

```python
## predict on whole test data

y_pred = model.predict(scaled_X_test)
y_pred
```

```python
y_test.iloc[:5]
```

```python
## 5. Evaluate

from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)

print(f"{mae}")
print(f"{mse}")
print(f"{rmse}")
```

```python
# Suppose you want to test: TV=100, Radio=50, Newspaper=20
my_sample = np.array([[100, 50, 20]])

# Scale it with the same scaler you fitted on training data
scaled_my_sample = scaler.transform(my_sample)

# Predict with the trained model
prediction = model.predict(scaled_my_sample)
print("Predicted Sales:", prediction[0])
```

```python
# Predict all test samples
y_pred = model.predict(scaled_X_test)

# Show first 10 actual vs predicted
for actual, pred in zip(y_test[:10], y_pred[:10]):
    print(f"Actual: {actual:.2f}, Predicted: {pred:.2f}")
```
